## Tensorflow Probability

This notebook is in progress to find out what it has to offer.

```python
import tensorflow as tf
tf.enable_eager_execution()
import tensorflow_probability as tfp
```

```python
from matplotlib import pyplot as plt
```

```python
tfd = tfp.distributions
d = tfd.Normal(loc=[-1.,1.], scale=1.)
x = d.sample(1000)
px = d.prob(x)
cdf = d.cdf(x)
print(d.mean())
print(d.variance())
print(d.mode())

print(d.entropy())
```

```python
plt.scatter(x[:,0], x[:,1])
plt.show()
plt.scatter(px[:,0], px[:,1])
plt.show()
plt.scatter(cdf[:,0], cdf[:,1])
plt.show()
```
